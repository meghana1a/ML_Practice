### Import Dataset

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix, accuracy_score, precision_score, recall_score
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("Iris.csv")
data
```

### Feature Selection


```python
selected_cols=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
x=data[selected_cols]
y=data.Species
```

```python
feature_names = x
labels = y.unique()
feature_names
```

## Spliting the dataset


```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

## Decision Tree Classifier


```python
clf = DecisionTreeClassifier(max_depth =4, random_state = 42)
clf.fit(X_train, y_train)
pred_decision_tree = clf.predict(X_test)
```

### Plot the Decision Tree using tree plot

```python
from sklearn import tree
import matplotlib.pyplot as plt

plt.figure(figsize=(30,10))

a = tree.plot_tree(clf,

                   class_names=labels,

                   rounded = True,

                   filled = True,

                   fontsize=14)

plt.show()
```

### Confusion Matrix


```python
cnf_matrix = confusion_matrix(y_test, pred_decision_tree )
print(cnf_matrix)
```

```python
cm_display = ConfusionMatrixDisplay(confusion_matrix = cnf_matrix)
cm_display.plot()
plt.show()
```

```python
print("Precision Score : ",precision_score(y_test, pred_decision_tree , average='micro'))
print("Recall Score : ",recall_score(y_test, pred_decision_tree , average='micro'))
print("Accuracy Score : ",accuracy_score(y_test, pred_decision_tree ))
```
